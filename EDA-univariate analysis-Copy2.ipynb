summary of data

* training set has 13,647,309 rows, 24 features and 24 response variables. 

Categorical Variables:
* fecha_dato
* ind_empleado: Employee index: A active, B ex employed, F filial, N not employee, P pasive
* pais_residencia: Customer's Country residence
* sexo: Customer'sex
* fecha_alta: The date in which the customer became as the first holder of a contract in the bank
* antiguedad: Customer seniority (in months)
* ult_fec_cli_1t: Last date as primary customer (if he isn't at the end of the month)

* canal_entrada: channel used by the customer to join
* indfall: Deceased index. N/S
* cod_prov: Province code (customer's address)
* nomprov: Province name
* segmento: segmentation: 01 - VIP, 02 - Individuals 03 - college graduated


Numerical Variables:
* ncodpers: Customer code
* age
* ind_nuevo: New customer Index. 1 if the customer registered in the last 6 months.
* indrel: 1 (First/Primary), 99 (Primary customer during the month but not at the end of the month)
* tipodom: Addres type. 1, primary address
* ind_actividad_cliente: Activity index (1, active customer; 0, inactive customer)
* renta: Gross income of the household

```python
def week_of_month(tgtdate):
    tgtdate = tgtdate.to_datetime()

    days_this_month = calendar.mdays[tgtdate.month]
    for i in range(1, days_this_month):
        d = datetime.datetime(tgtdate.year, tgtdate.month, i)
        if d.day - d.weekday() > 0:
            startdate = d
            break
    # now we canuse the modulo 7 appraoch
    return (tgtdate - startdate).days //7 + 1
```

```python
def clean_data(df):
    df['age'] = pd.to_numeric(df.age, errors='coerce')
    df['cod_prov'] = df['cod_prov'].astype('object')
    # delete rows whose columns have 27734 missing vlaues. 
    df.dropna(subset=['ind_empleado', 'pais_residencia'], how='all', inplace = True)
    # add new level for categorical data 
    df[['sexo', 'indrel_1mes', 'tiprel_1mes', 'conyuemp', 'canal_entrada', 'segmento', 'cod_prov', 'nomprov']]\
    = df[['sexo', 'indrel_1mes', 'tiprel_1mes','conyuemp','canal_entrada', 'segmento', 'cod_prov', 'nomprov']].fillna('U')
    if df[df.renta.isnull()].shape[0] != 0:
    # use median to fill out numerical variable: household income
        df['renta'].fillna((df['renta'].median()), inplace=True)
    # fill out the last day of primary customers. NAs represent these customers are still primary customers. 
    df['ult_fec_cli_1t'] = df['ult_fec_cli_1t'].fillna('2020-06-28')
    # combine levels together 
    df.indrel_1mes[(df.indrel_1mes == '1.0') | (df.indrel_1mes == '1')] = 1.0
    df.indrel_1mes[(df.indrel_1mes == '2.0') | (df.indrel_1mes == '2')] = 2.0
    df.indrel_1mes[(df.indrel_1mes == '3.0') | (df.indrel_1mes == '3')] = 3.0
    df.indrel_1mes[(df.indrel_1mes == '4.0') | (df.indrel_1mes == '4')] = 4.0
    # two response variables have NAs, not much. So delete them
    if str('ind_nomina_ult1') in(df.columns):
        df.dropna(subset=['ind_nomina_ult1', 'ind_nom_pens_ult1'], how='all', inplace = True)
    # all customers have the same value: 1, not providing information
    df.drop(columns = ['tipodom'] , inplace = True)
    # more than 99% of NAs
    df.drop(columns = ['conyuemp'] , inplace = True)
    # 4% of age greater than 75 years old(1.5 times of upper quartile)
    # df = df[df['age'] <= 75]
    df.loc[df.age > 75,"age"] = 83 #83 is median age of people older than 75 years old
    df['fecha_alta'] = pd.to_datetime(df['fecha_alta'])
    # monday is 0; sunday is 6
    df['custoemr_dow'] = df['fecha_alta'].dt.dayofweek
    df['customer_wom'] = df['fecha_alta'].apply(week_of_month)
    df['customer_m'] = pd.DatetimeIndex(df['fecha_alta']).month
    df['customer_y'] = pd.DatetimeIndex(df['fecha_alta']).year
    df['ult_fec_cli_1t'] = pd.to_datetime(df['ult_fec_cli_1t'])
    df['lastprimary_dow'] = df['ult_fec_cli_1t'].dt.dayofweek
    df['lastprimary_wom'] = df['ult_fec_cli_1t'].apply(week_of_month)
    df['lastprimary_m'] = pd.DatetimeIndex(df['ult_fec_cli_1t']).month
    df['lastprimary_y'] = pd.DatetimeIndex(df['ult_fec_cli_1t']).year
    return df
```

```python
def plot_binary(data, col_name, title):
    p1 = sns.countplot(x=col_name, data=data, palette="Greens_d")
    plt.title(title, fontsize = 15)
    for p in p1.patches:
        height = p.get_height()
        p1.text(p.get_x()+p.get_width()/2.,
                height + 3,
                '{:1.2f}'.format(height/data.shape[0]),
                ha="center") 

    plt.show()
```

```python
def plot_df(df, colname, column_name, groupby_name, groupby_name_list): 
    result = df.groupby([groupby_name, colname], as_index = False)['ncodpers'].count()
    result_0 = result[result[groupby_name] == 0][[ colname,'ncodpers']]
    result_1 = result[result[groupby_name] == 1][[ colname,'ncodpers']]
    result_0.columns = [column_name, str(product[0]) +'_num' ]
    result_1.columns = [column_name, str(product[0]) +'_num' ]


    for name in groupby_name_list:
        a = df.groupby([name, colname], as_index = False)['ncodpers'].count()
        a_0 = a[a[name] == 0][[colname, 'ncodpers']]
        a_1 = a[a[name] == 1][[colname, 'ncodpers']]
        a_0.columns = [column_name, str(name) +'_num' ]
        a_1.columns = [column_name, str(name) +'_num' ]
        result_0 = pd.merge(result_0, a_0 , how = 'left', on = [column_name])
        result_1 = pd.merge(result_1, a_1 , how = 'left', on = [column_name])
        result_0T = result_0.transpose()
        result_0T.columns = result_0T.iloc[0]
        result_0T = result_0T.iloc[1:]
        result_1T = result_1.transpose()
        result_1T.columns = result_1T.iloc[0]
        result_1T = result_1T.iloc[1:]
    return result_0T, result_1T
```
